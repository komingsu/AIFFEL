```python
import pandas as pd

from sklearn.datasets import load_breast_cancer
from sklearn.datasets import load_wine
from sklearn.datasets import load_digits

from sklearn.model_selection import train_test_split
from sklearn.metrics import balanced_accuracy_score, mean_squared_error, confusion_matrix, recall_score, r2_score

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import SGDClassifier, LogisticRegression

import warnings
warnings.filterwarnings(action='ignore')
```

# 유방암 진단

유방암 여부를 판단하는 머신러닝 구현

```python
cancer_data = load_breast_cancer()



X_train, X_test, y_train, y_test = train_test_split(cancer_data.data, 
                                                    cancer_data.target, 
                                                    test_size=0.2) 


print("X_train.shape :", X_train.shape,"X_test.shape :", X_test.shape,"Y_train.shape :", y_train.shape,"Y_test.shape :", y_test.shape, sep="  ")
```

```python
cancer_df = pd.DataFrame(cancer_data.data)
cancer_df.columns = cancer_data.feature_names
cancer_df["target"] = cancer_data.target
cancer_df["target"] = cancer_df["target"].astype(str)

display(cancer_df)
display(cancer_df.describe())

print("target count: " , cancer_df["target"].value_counts(), sep="\n")
```

## model

```python
decision_tree = DecisionTreeClassifier()
random_forest = RandomForestClassifier()
svm_model = SVC()
sgd_model = SGDClassifier()
logistic_model = LogisticRegression()

decision_tree.fit(X_train, y_train)
random_forest.fit(X_train, y_train)
svm_model.fit(X_train, y_train)
sgd_model.fit(X_train, y_train)
logistic_model.fit(X_train, y_train);
```

## predict

```python
dt_y_pred = decision_tree.predict(X_test)
rf_y_pred = random_forest.predict(X_test)
svm_y_pred = svm_model.predict(X_test)
sgd_y_pred = sgd_model.predict(X_test)
lr_y_pred = logistic_model.predict(X_test)
```

```python
print("decision_tree :")
print(confusion_matrix(y_test, dt_y_pred))
print(f"balanced_accuracy_score : {balanced_accuracy_score(y_test, dt_y_pred):.4f},\t recall_score : {recall_score(y_test, dt_y_pred):.4f}")
print("\n random_forest :")
print(confusion_matrix(y_test, rf_y_pred))
print(f"balanced_accuracy_score : {balanced_accuracy_score(y_test, rf_y_pred):.4f},\t recall_score : {recall_score(y_test, rf_y_pred):.4f}")
print("\n svm :")
print(confusion_matrix(y_test, svm_y_pred))
print(f"balanced_accuracy_score : {balanced_accuracy_score(y_test, svm_y_pred):.4f},\t recall_score : {recall_score(y_test, svm_y_pred):.4f}")
print("\n SGDClassifier :")
print(confusion_matrix(y_test, sgd_y_pred))
print(f"balanced_accuracy_score : {balanced_accuracy_score(y_test, sgd_y_pred):.4f},\t recall_score : {recall_score(y_test, sgd_y_pred):.4f}")
print("\n LogisticRegression :")
print(confusion_matrix(y_test, lr_y_pred))
print(f"balanced_accuracy_score : {balanced_accuracy_score(y_test, lr_y_pred):.4f},\t recall_score : {recall_score(y_test, lr_y_pred):.4f}")
```

암을 진단하는 경우에는 음성을 양성으로 진단하면 안되기 때문에 비록 accuarcy_score는 random_forest 가 더 좋지만 SGDClassifier를 채택하여야 할것같습니다.
